```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 读取数据并展示
GGbond = plt.imread('./GGbond_1.jpg')
plt.imshow(GGbond)
```

```python
# 左右翻转并显示
GGbond_1 = tf.image.random_flip_left_right(GGbond)
plt.imshow(GGbond_1)
```

```python
# 上下翻转
GGbond_2 = tf.image.random_flip_up_down(GGbond)
plt.imshow(GGbond_2)
```

```python
# 随机裁剪
GGbond_3 = tf.image.random_crop(GGbond, (200, 200, 3))
plt.imshow(GGbond_3)
```

# #  颜色变换


```python
# 亮度随机变化为原图亮度的50% (增强或减弱)
GGbond_4 = tf.image.random_brightness(GGbond, 0.6)
plt.imshow(GGbond_4)
```

```python
# 随机变化图像的色调
GGbond_5 = tf.image.random_hue(GGbond, 0.5)
plt.imshow(GGbond_5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from keras.preprocessing.image import ImageDataGenerator
```

```python
# 获取数据集
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
# 将数据转换为4维的形式
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
```

```python
# 设置图像增强方式：水平翻转
datagen = ImageDataGenerator(horizontal_flip=True)
```

```python
# 查看增强后的结果
for x_batch, y_batch in datagen.flow(x_train, y_train, batch_size=9):
    plt.figure(figsize=(8, 8))  # 设定每个图像显示的大小
    # 产生一个3*3网格的图像
    for i in range(0, 9):
        plt.subplot(330 + 1 + i)
        plt.title(y_batch[i])
        plt.axis('off')
        plt.imshow(x_batch[i].reshape(28, 28), cmap='gray')
    plt.show()
    break
```
